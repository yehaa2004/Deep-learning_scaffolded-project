```python
# ============================================================
# PRETRAINED CNN COMPARISON


import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
from datasets import load_dataset
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
from sklearn.utils.class_weight import compute_class_weight

# ============================================================
# REPRODUCIBILITY
# ============================================================

SEED=42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

print("\n=========== PROBLEM FORMULATION ===========")
print("Marine Animal Image Classification using Transfer Learning")

# ============================================================
# LOAD DATASET
# ============================================================

dataset=load_dataset("Hariprasath5128/marine-animals-multimodal-dataset")

dataset=dataset["train"].train_test_split(test_size=0.3,seed=SEED)
temp=dataset["test"].train_test_split(test_size=0.5,seed=SEED)

train_data=dataset["train"]
val_data=temp["train"]
test_data=temp["test"]

print("\nDataset Split: 70% Train / 15% Validation / 15% Test")

# ============================================================
# PREPROCESSING
# ============================================================

IMG_SIZE=96
BATCH_SIZE=8

def preprocess(example):

    img=example["image"].convert("RGB")
    img=img.resize((IMG_SIZE,IMG_SIZE))
    img=np.array(img)/255.0

    return img,example["label"]

def generator(data):

    for item in data:
        yield preprocess(item)

num_classes=len(set(train_data["label"]))

train_ds=tf.data.Dataset.from_generator(
    lambda:generator(train_data),
    output_signature=(
        tf.TensorSpec(shape=(IMG_SIZE,IMG_SIZE,3),dtype=tf.float32),
        tf.TensorSpec(shape=(),dtype=tf.int64)
    )
).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

val_ds=tf.data.Dataset.from_generator(
    lambda:generator(val_data),
    output_signature=(
        tf.TensorSpec(shape=(IMG_SIZE,IMG_SIZE,3),dtype=tf.float32),
        tf.TensorSpec(shape=(),dtype=tf.int64)
    )
).batch(BATCH_SIZE)

test_ds=tf.data.Dataset.from_generator(
    lambda:generator(test_data),
    output_signature=(
        tf.TensorSpec(shape=(IMG_SIZE,IMG_SIZE,3),dtype=tf.float32),
        tf.TensorSpec(shape=(),dtype=tf.int64)
    )
).batch(BATCH_SIZE)

print("\nNormalization: Pixel scaling [0,1]")

# ============================================================
# CLASS IMBALANCE HANDLING
# ============================================================

labels=train_data["label"]

class_weights=compute_class_weight(
    class_weight="balanced",
    classes=np.unique(labels),
    y=labels)

class_weights=dict(enumerate(class_weights))

# ============================================================
# MODEL BUILDER
# ============================================================

def build_model(model_name,lr):

    if model_name=="MobileNetV2":

        base=tf.keras.applications.MobileNetV2(
            input_shape=(IMG_SIZE,IMG_SIZE,3),
            include_top=False,
            weights="imagenet",
            alpha=0.35)

    if model_name=="EfficientNetB0":

        base=tf.keras.applications.EfficientNetB0(
            input_shape=(IMG_SIZE,IMG_SIZE,3),
            include_top=False,
            weights="imagenet")

    base.trainable=False

    inputs=tf.keras.Input(shape=(IMG_SIZE,IMG_SIZE,3))

    x=base(inputs,training=False)

    x=tf.keras.layers.GlobalAveragePooling2D()(x)

    x=tf.keras.layers.Dense(128,activation="relu")(x)

    x=tf.keras.layers.Dropout(0.5)(x)

    outputs=tf.keras.layers.Dense(num_classes,activation="softmax")(x)

    model=tf.keras.Model(inputs,outputs)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"])

    return model,base

# ============================================================
# HYPERPARAMETER EXPERIMENTS
# ============================================================

learning_rates=[0.001,0.0005]

models=["MobileNetV2","EfficientNetB0"]

results={}

for name in models:

    for lr in learning_rates:

        print("\n==============================")
        print("Training:",name," LR:",lr)
        print("==============================")

        model,base=build_model(name,lr)

        model.summary()

        history=model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=4,
            class_weight=class_weights)

        # ============================================================
        # FINE TUNING
        # ============================================================

        print("\nFine tuning last layers")

        base.trainable=True

        for layer in base.layers[:-20]:
            layer.trainable=False

        model.compile(
            optimizer=tf.keras.optimizers.Adam(lr/10),
            loss="sparse_categorical_crossentropy",
            metrics=["accuracy"])

        model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=2)

        # ============================================================
        # TEST EVALUATION
        # ============================================================

        y_true=[]
        y_pred=[]

        for images,labels in test_ds:

            preds=model.predict(images)

            y_true.extend(labels.numpy())
            y_pred.extend(np.argmax(preds,axis=1))

        acc=accuracy_score(y_true,y_pred)
        prec=precision_score(y_true,y_pred,average="weighted")
        rec=recall_score(y_true,y_pred,average="weighted")
        f1=f1_score(y_true,y_pred,average="weighted")

        results[(name,lr)]=[acc,prec,rec,f1]

        print("\nAccuracy:",acc)
        print("Precision:",prec)
        print("Recall:",rec)
        print("F1 Score:",f1)

        # ============================================================
        # CONFUSION MATRIX
        # ============================================================

        cm=confusion_matrix(y_true,y_pred)

        plt.figure(figsize=(8,6))
        plt.imshow(cm,cmap="Blues")
        plt.title("Confusion Matrix")
        plt.colorbar()
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()

        # ============================================================
        # LEARNING CURVES
        # ============================================================

        plt.plot(history.history["loss"])
        plt.plot(history.history["val_loss"])
        plt.title("Loss Curve")
        plt.legend(["Train","Validation"])
        plt.show()

        plt.plot(history.history["accuracy"])
        plt.plot(history.history["val_accuracy"])
        plt.title("Accuracy Curve")
        plt.legend(["Train","Validation"])
        plt.show()

# ============================================================
# MODEL COMPARISON
# ============================================================

print("\n=========== MODEL COMPARISON ===========")

for k,v in results.items():

    print("\nModel:",k[0]," LR:",k[1])
    print("Accuracy:",round(v[0],4))
    print("Precision:",round(v[1],4))
    print("Recall:",round(v[2],4))
    print("F1:",round(v[3],4))

# ============================================================
# BIAS VARIANCE ANALYSIS
# ============================================================

train_acc=history.history["accuracy"][-1]
val_acc=history.history["val_accuracy"][-1]

print("\n=========== BIAS VARIANCE ANALYSIS ===========")

print("Train Accuracy:",train_acc)
print("Validation Accuracy:",val_acc)

if abs(train_acc-val_acc)<0.05:
    print("Balanced bias-variance tradeoff")
elif train_acc>val_acc:
    print("Possible overfitting")
else:
    print("Good generalization")

print("\nFeature representation: CNN convolutional layers extract spatial features.")
print("Pooling layers reduce dimensionality.")
print("Transfer learning enables faster convergence.")

print("\nTotal Parameters:",model.count_params())
```

```python
# ============================================================
# TEST EVALUATION (FIXED)
# ============================================================

y_true = []
y_pred = []

for images, labels in test_ds:

    preds = model.predict(images, verbose=0)

    preds = np.argmax(preds, axis=1)

    y_pred.extend(preds)
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# ============================================================
# METRICS
# ============================================================

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="weighted", zero_division=0)
rec = recall_score(y_true, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)

print("\n=========== PERFORMANCE ===========")
print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
print("F1 Score :", f1)


# ============================================================
# CONFUSION MATRIX (FIXED VISUALIZATION)
# ============================================================

import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12,10))

sns.heatmap(
    cm,
    cmap="Blues",
    xticklabels=range(num_classes),
    yticklabels=range(num_classes)
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.show()
```
